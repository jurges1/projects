```python
weather = pd.read_csv('weather.csv')
```

```python
#there are cases when you have a lot of data when the information is repeated and in a row you will have a lot of data that are repeated except of few numbers 
# that's how you know that variables are actually stored in a row 
#so in this case we have duplicate rows so we need to do the oposite of melting and take two columns for many rows and unpivot them

weather
```

```python
#first thing  to do is to melt the day columns into one column in order to  shrink the dataset size and aggregate better

weather_melt = pd.melt(weather,id_vars=['id','year','month','element'],var_name='day',value_name='temp')
weather_melt.head()
```

```python
# in order too make a column into a row we use pivot which turns the value of the row in columns
#we have two methods .pivot and .pivot_table. if you use pivot you make assumption that there are no duplicates
```

```python
#so this function does this. index is for columns you don't want to change, in the column section is the column 
# you want to divide for each unique value, and the temp is the value you want to aggregate for column section, dropna is for dropping mising values

weather_tidy = weather_melt.pivot_table(
index= ['id','year','month','day'],
columns='element',
values='temp'
)
```

```python
weather_tidy
```

```python
#the format that is printed above it shows that there are some irregulaties in the way the dataframe is shown so we can use reset_index and it will flat everything down

weather_tidy.reset_index()
```
